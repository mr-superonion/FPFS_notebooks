```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
import gc
import fitsio
import numpy as np
from fpfs import pltutil
import matplotlib.pylab as plt
from matplotlib import container

import matplotlib as mpl
mpl.rcParams['axes.linewidth'] = 2.0 #set the value globally
```

```python
plt.close()
fig,axes=pltutil.make_figure_axes(ny=2,nx=1,square=False)
ax=axes[0]
xlabel=r'$m_{\rm F}$ upper limit';
ax.set_ylabel('$m$ $[10^{-2}]$',fontsize=18)
ax.set_yscale('symlog',linthresh=1.)
ax.set_yticks([-5,-1,-0.5,0.,0.5,1,5])
ax.set_yticklabels(['-5','-1','-0.5','0.','0.5','1','5'])
ax.set_xlim(21.8,25.6)
ax.axvline(25.,color='black',ls='-.')
ax.set_ylim(-6,6.)
ax.invert_xaxis()
ax.tick_params(width=2,size=4)
ax.set_xticklabels([])
ax.grid()
ax.spines['bottom'].set_visible(False)
ax.axhspan(-0.3,0.3,0,100,color='gray',alpha=0.5)
ax.axhspan(-1.,1.,0,100,color='gray',alpha=0.2)

ax2=axes[1]
ax2.set_ylabel('$c$ $[10^{-4}]$',fontsize=18)
ax2.set_xlim(21.8,25.6)
ax2.axvline(25.,color='black',ls='-.')
ax2.set_ylim(-2.2,2.5)
ax2.tick_params(width=2,size=4)
ax2.set_xlabel(xlabel,fontsize=18)
ax2.grid()
ax2.invert_xaxis()
plt.tight_layout(h_pad=-0.2)

outs=fitsio.read('../3_blended_min32_newdet_try2/detect_magcut.fits')
ave=np.average(outs,axis=0)
std=np.std(outs,axis=0)
x=ave[0,:]
y1=ave[1,:]/ave[4,:]/0.04-1
y2=ave[1,:]/ave[5,:]/0.04-1
z1=ave[2,:]/ave[4,:]
z2=ave[3,:]/ave[5,:]
yerr=std[1,:]/ave[5,:]/0.04/np.sqrt(len(outs))
zerr=std[3,:]/ave[5,:]/np.sqrt(len(outs))
ax.errorbar(np.array(x)+0.039,np.array(y1)*1e2,np.array(yerr)*1e2,color=pltutil.colors[1])
ax.errorbar(np.array(x)+0.013,np.array(y2)*1e2,np.array(yerr)*1e2,color=pltutil.colors[1],ls='--')
ax2.errorbar(np.array(x)+0.039,np.array(z1)*1e4,np.array(zerr)*1e4,label='$\sigma_h=$0."59',color=pltutil.colors[1])
ax2.errorbar(np.array(x)+0.013,np.array(z2)*1e4,np.array(zerr)*1e4,color=pltutil.colors[1],ls='--')

outs=fitsio.read('../3_blended_min32_newdet_try3/detect_magcut.fits')
ave=np.average(outs,axis=0)
std=np.std(outs,axis=0)
x=ave[0,:]
y1=ave[1,:]/ave[4,:]/0.04-1
y2=ave[1,:]/ave[5,:]/0.04-1
z1=ave[2,:]/ave[4,:]
z2=ave[3,:]/ave[5,:]
yerr=std[1,:]/ave[5,:]/0.04/np.sqrt(len(outs))
zerr=std[3,:]/ave[5,:]/np.sqrt(len(outs))
ax.errorbar(np.array(x)-0.013,np.array(y1)*1e2,np.array(yerr)*1e2,color=pltutil.colors[2])
ax.errorbar(np.array(x)-0.039,np.array(y2)*1e2,np.array(yerr)*1e2,color=pltutil.colors[2],ls='--')
ax2.errorbar(np.array(x)-0.013,np.array(z1)*1e4,np.array(zerr)*1e4,label='$\sigma_h=$0."45',color=pltutil.colors[2])
ax2.errorbar(np.array(x)-0.039,np.array(z2)*1e4,np.array(zerr)*1e4,color=pltutil.colors[2],ls='--')

handles, labels = ax2.get_legend_handles_labels()
handles = [h[0] if isinstance(h, container.ErrorbarContainer) else h for h in handles]
leg=ax2.legend(handles, labels,fontsize=20,handlelength=0,handletextpad=0,fancybox=False,markerscale=0)
for item in leg.legendHandles:
    item.set_visible(False)
for it,text in enumerate(leg.get_texts()):
    cc=leg.legendHandles[it].get_color()
    text.set_color(cc)
    
plt.tight_layout(h_pad=-0.2)
fig.savefig('blend_magcut_all.pdf')
```
